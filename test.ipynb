```python
import matplotlib.pyplot as plt
import torchvision

# 下载并加载MNIST数据集
"""
使用 torchvision.transforms.ToTensor() 转换图像时，图像的像素值会被归一化到 [0, 1] 范围的浮点数。
"""
transform = torchvision.transforms.ToTensor()
mnist_data = torchvision.datasets.MNIST(root='data', train=True, transform=transform, download=True)

# 获取第一个图像及其对应的标签
image, label = mnist_data[0]

# 因为MNIST图像是灰度图，只有一个通道，使用squeeze去掉不必要的维度
image = image.squeeze()

# 展示图像
plt.imshow(image, cmap='gray')
plt.title(f'Label: {label}')
plt.show()
```

```python
import matplotlib.pyplot as plt
import torchvision

# 下载并加载MNIST数据集
transform = torchvision.transforms.ToTensor()
mnist_data = torchvision.datasets.MNIST(root='data', train=True, transform=transform, download=True)

# 获取第一个图像及其对应的标签
image, label = mnist_data[0]

# 去掉不必要的维度，使其成为 (28, 28) 的二维数组
image = image.squeeze()

# 创建图形和子图
fig, ax = plt.subplots()
ax.imshow(image, cmap='gray')

# 在每个像素上标注其数值
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        pixel_value = image[i, j].item()  # 获取像素值
        print(pixel_value)
#         ax.text(j, i, f'{pixel_value:.2f}', ha='center', va='center', color='red', fontsize=6)
# 
# # 设置标题
# ax.set_title(f'Label: {label}')
# plt.show()
```

```python
import torch
import torch.nn as nn

# 定义交叉熵损失函数
lossF = nn.CrossEntropyLoss()

# 模拟的模型输出（logits），假设batch_size=3, num_classes=3
outputs = torch.tensor([[2.0, 1.0, 0.1],
                        [0.5, 2.5, 0.3],
                        [2.2, 0.1, 1.3]])

# 模拟的真实标签
labels = torch.tensor([0, 1, 2])  # 每个样本对应一个类别

# 计算交叉熵损失
loss = lossF(outputs, labels)

print(loss.item())  # 输出损失值
```
